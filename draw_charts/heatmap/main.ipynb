```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
df = pd.read_csv("./COMPOSED.csv")

plt.figure(figsize=(16,10))
sns.pairplot(df, kind="reg")
```

```python
cmap = "YlGnBu"
plt.figure(figsize=(16,10))
relation = df.corr()
sns.heatmap(relation,xticklabels=relation.columns,yticklabels=relation.columns,annot=True,cmap=cmap,center=0,annot_kws={"fontsize": 12,"fontweight":"demibold"})
```
