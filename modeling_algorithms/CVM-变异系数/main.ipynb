# Coefficient of Variation 变异系数

`变异系数`，可以体现对象数据与标准数据的相对差异,数值越大，表示不平衡程度越大。
其定义为`标准差`$\sigma$与`平均值`$\mu$之比。

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("./data/average-per-capita-supply-of-pulses.csv")
df = df[df.Code.isnull() == False] # 去掉国家代码Code为空的行，也就是大洲和组织
df = df.reset_index(drop=True)
df = df.drop(columns=["Entity"])
df.columns=["Country","Area","Value"] # 把年当成地区把，这只是为了演示
df
```

## 求出不同地区的变异系数

注意看上图表格，把年份当成地区来看吧，只是演示

```python
Countries = df["Country"].unique()
res = dict()
for Country in Countries:
    tmp = df[df["Country"] == Country]
    mean = tmp["Value"].mean()
    std = tmp["Value"].std()
    cv = std/mean
    res[Country] = cv
print(res)
```
